```python
# implement load method
def load(f, label):
    # load the file into tensor
    image = tf.io.read_file(f)
    
    # decode it to JPEG format
    image = tf.image.decode_jpeg(image)
    
    # convert it to tf.float32
    image = tf.cast(image, tf.float32)
    
    return image, label
```

```python
class Cnn(nn.Module):
    def __init__(self):
        super(Cnn, self).__init__()
        
        self.layer1 = nn.Sequential(
            nn.Conv2d(3, 16, kernel_size=3, padding=0, stride=2),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        
        self.layer2 = nn.Sequential(
            nn.Conv2d(16, 32, kernel_size=3, padding=0, stride=2),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        
        self.layer3 = nn.Sequential(
            nn.Conv2d(32, 64, kernel_size=3, padding=0, stride=2),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        
        # Calculate the number of features after the last convolutional layer
        self.num_features = self._get_conv_output((3, 224, 224))
        
        self.fc1 = nn.Linear(self.num_features, 128) 
        self.dropout = nn.Dropout(0.5)
        self.fc2 = nn.Linear(128, 2)
        self.relu = nn.ReLU()
        
    def _get_conv_output(self, shape):
        """
        Utility function to compute the output size of convolutional layers.
        """
        batch_size = 1
        input = torch.autograd.Variable(torch.rand(batch_size, *shape))
        output_feat = self._forward_features(input)  # Forward pass to get the output size
        n_size = output_feat.data.view(batch_size, -1).size(1)
        return n_size

    def _forward_features(self, x):
        """
        Utility function to perform forward pass until the last convolutional layer.
        """
        x = self.layer1(x)
        x = self.layer2(x)
        x = self.layer3(x)
        return x
        
    def forward(self, x):
        out = self._forward_features(x)
        out = out.view(out.size(0), -1)
        out = self.relu(self.fc1(out))
        out = self.dropout(out)
        out = self.fc2(out)
        return out
```

```python
model = Cnn().to(device)
model.train()
```

```python
#set loss function and optimizer
optimizer = optim.RMSprop(params = model.parameters(),lr=0.02)
criterion = nn.CrossEntropyLoss()
```

```python
#train the network
epochs = 20
count = 0
loss_list = []
accuracy_list = []
iteration_list = []

for epoch in range(epochs):
    epoch_loss = 0
    epoch_accuracy = 0
    
    for data, label in train_loader:
        data = data.to(device)
        label = label.to(device)
        
        output = model(data)
        loss = criterion(output, label)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        acc = ((output.argmax(dim=1) == label).float().mean())
        epoch_accuracy += acc/len(train_loader)
        epoch_loss += loss/len(train_loader)
        
    count += 1
    accuracy_list.append(epoch_accuracy)
    loss_list.append(epoch_loss)
    iteration_list.append(count)
        
    print('Epoch : {}, train accuracy : {}, train loss : {}'.format(epoch+1, epoch_accuracy,epoch_loss))
```

```python
len(iteration_list)
new_loss_list = []
new_accuracy_list = []


for i in range(len(iteration_list)):
    new_loss_list.append(loss_list[i].detach().numpy())
    new_accuracy_list.append(accuracy_list[i].detach().numpy())
```

```python
#visualization loss
graph.plot(iteration_list,new_loss_list)
graph.xlabel("Number of Iterations")
graph.ylabel("Loss")
graph.title("CNN: Loss vs Number of iteration")
graph.grid(True)
graph.show()

#Visualization accuracy
graph.plot(iteration_list,new_accuracy_list, color ="red")
graph.xlabel("number of iterations")
graph.ylabel("Accuracy")
graph.title("CNN: Accuracy vs Number of iteration")
graph.grid(True)
graph.show()
```

```python
#TESTING AND EVALUATING THE MODEL
import torch.nn.functional as F
pneumonia_probs = []
model.eval()
with torch.no_grad():
    for data, fileid in test_loader:
        data = data.to(device)
        preds = model(data)
        preds_list = F.softmax(preds, dim=1)[:, 1].tolist()
        pneumonia_probs += list(zip(list(fileid), preds_list))
        
        
```
